<a id='Slicing And Dicing  OF Dataframe'></a>
### Slicing And Dicing  OF Dataframe

```python
# BASE DATA
data = pd.read_csv("enrollments.csv")
tempdata = data.head(n =7)

print("COLUMN SELECTION WITH NAME : \n{}".format(tempdata[['status','is_canceled']]))
# Columns has to be in [] if there are more than one columns to display.
# For single column [] is not required(tempdata['status'])
print("____________________________________________________________________________________________")
print("ROW SELECTION WITH Indexes : \n")
print("ROW SELECTION WITH loc: \n{}".format(tempdata.loc[[1,2]])) # loc['abc','xyz'], if abc and xyz were indexes
print("ROW SELECTION WITH iloc : \n{}".format(tempdata.iloc[[1,2]]))
print("ROW SELECTION WITH range : \n{}".format(tempdata.loc[1:2]))
#All above 3 returns the same results. As the indexes are sequential numbers here,loc and iloc will return same values.
#In row selection also multiple rows require []. Simgle row and range does not need this
print("____________________________________________________________________________________________")
print("ROW SELECTION WITH STEP: \n{}".format(tempdata.loc[0::3])) # start = 0, stop = end , step = 2
print("____________________________________________________________________________________________")
print("ROW/COLUMN SELECTION: \n")
print("ROW/COLUMN SELECTION: \n{}".format(tempdata.loc[1:2][['status','is_canceled']]))
print("ROW/COLUMN SELECTION: \n{}".format(tempdata.loc[1:2, ['status','is_canceled']])) # loc[[rows],[columns]]
print("ROW/COLUMN SELECTION: \n{}".format(tempdata.iloc[1:3,1:4])) #iloc should have index numbers for row/column
```

```python
# BASE DATA
data = pd.read_csv("enrollments.csv")
tempdata = data.head(n =7)

print("____________________________________________________________________________________________")
print("COLUMN SELECTION USING FILTER COMMAND (row with is_canceled in True): \n")
print("ITEMS :\n{}".format(tempdata.filter(items=['account_key','status'],axis ='columns')))
print("REGEX :\n{}".format(tempdata.filter(regex ='date$',axis ='columns')))
print("REGEX :\n{}".format(tempdata.filter(regex ='^is',axis ='columns')))
print("LIKE :\n{}".format(tempdata.filter(like ='da',axis ='columns')))

print("____________________________________________________________________________________________")
print("ROW SELECTION USING FILTER COMMAND (row with is_canceled in True): \n")
print("ITEMS :\n{}".format(tempdata.filter(items=[1,2,3],axis ='index')))
# REGEX and LIKE Doesnot work here as the indexes are numeric
```

<a id='Filtering'></a>
### Filtering

```python
# BASE DATA
data = pd.read_csv("enrollments.csv")
tempdata = data.head(n =5)

print("____________________________________________________________________________________________")
print("FILTER (row with is_canceled in True): \n{}".format(tempdata[tempdata['is_canceled']==True]))
# tempdata['is_canceled']==True] => returns a Boolean dataframe with same number of rows as that parent dataset 
print("____________________________________________________________________________________________")
print("FILTER WITH SELECTED COLUMN(days_to_cancel == 0)\n")
print("Status :\n{}".format(tempdata[tempdata['days_to_cancel']== 0]['status']))
print("Status :\n{}".format(tempdata[tempdata['days_to_cancel']== 0][['status','is_canceled']]))
print("____________________________________________________________________________________________")
print("FILTER WITH MULTIPLE CONDITIONS (days_to_cancel == 0)\n")
print("MULTIPLE FILTER AND :\n{}".format(tempdata[(tempdata['status']== 'canceled') & (tempdata['days_to_cancel']== 0)]))
print("MULTIPLE FILTER OR :\n{}".format(tempdata[(tempdata['status']== 'canceled') | (tempdata['days_to_cancel']== 0)]))
```

### Add/ Drop/ Rename Columns
#### Drop

```python
# BASE DATA
data = pd.read_csv("enrollments.csv")

print("DROP COLUMNS (use [] if there are more than 1 columns): \n")
print("BEFORE: \n{}".format(data.columns))
data.drop('status',axis = 'columns',inplace=True)
data.drop(columns = 'account_key',inplace=True)
print("AFTER: \n{}".format(data.columns))
print("____________________________________________________________________________________________")

print("DROP ROWS (use [] if there are more than 1 rows): \n")
print("BEFORE: \n{}".format(data.index))
data.drop(0,axis = 'index',inplace=True)
data.drop(index =[2,3,4],inplace=True)
print("AFTER: \n{}".format(data.index))
print("____________________________________________________________________________________________")

print("DROP BOTH ROWS AND COLUMNS (use [] if there are more than 1 rows): \n")
data.drop(index =7,columns ='is_canceled' ,inplace=True)
print(data.head())
```

<a id='Handling Null(NaN)'></a>
### Handling Null(NaN)
#### COUNT NaN and Fill Values for  NaN

```python
# BASE DATA

data = pd.DataFrame(
{
    'Name' : ['Mike' ,'Bill'  ,'Ram'   ,'Scot' ,'Rose' ,np.nan],
     'age' : [25     , np.nan , 27     , 30    , 21    ,np.nan],
     'Sex' : ['M'    , 'M'    , np.nan , 'M'   , 'F'   ,np.nan],
    'Blank': [np.nan , np.nan , np.nan ,np.nan ,np.nan ,np.nan ]
    
})

print("PRINT NUMBER OF NaN IN EACH COLUMN \n")
print("NaN columns: \n{}".format(data.isna().sum())) # isnull() also can be used insted of usna()
print("____________________________________________________________________________________________")
print("____________________________________________________________________________________________")

fillNaFill = data
print("FILL NaN ALL: \n")
print("BEFORE:")
display(fillNaFill)
fillNaFill= fillNaFill.fillna(0)
print("AFTER :")
display(fillNaFill)

print("____________________________________________________________________________________________")
print("____________________________________________________________________________________________")

fillNaOne = data
print("FILL NaN One Column: \n")
print("BEFORE:")
display(fillNaOne)
fillNaOne['age'] = fillNaOne['age'].fillna(25)
print("AFTER :")
display(fillNaOne)

print("____________________________________________________________________________________________")
print("____________________________________________________________________________________________")
fillNamany = data
print("FILL NaN With Differnt Values in each Column: \n")
print("BEFORE:")
display(fillNamany)

updateVal = {
    'Name' : 'dummy',
     'age' : 0,
     'Sex' : 'O',
    'Blank': 'Blank'
    
}
fillNamany = fillNamany.fillna(value  = updateVal)
print("AFTER :")
display(fillNamany)


```

#### Drop Rows With NaN

```python
dropRowData = pd.DataFrame(
{
    'Name' : ['Mike' ,'Bill'  ,'Ram'   ,'Scot' ,'Rose' ,np.nan],
     'age' : [25     , np.nan , 27     , 30    , 21    ,np.nan],
     'Sex' : ['M'    , 'M'    , np.nan , 'M'   , 'F'   ,np.nan],
    'Blank': [np.nan , np.nan , np.nan ,np.nan ,np.nan ,np.nan ]
    
})

print("DROP ROWS: \n")
print("BEFORE:")
display(dropRowData)
dropRowData.dropna(axis = 'index',how = 'all',inplace = True) 
print("AFTER 1: ROWS WITH ALL COLUMNS AS NaN :")
display(dropRowData)
dropRowData.dropna(axis = 'index',how = 'any',inplace = True) 
print("AFTER 2: ROWS WITH ALL COLUMNS AS NaN :")
display(dropRowData)
```

#### Drop Columns With NaN

```python
dropColumnData = pd.DataFrame(
{
    'Name' : ['Mike' ,'Bill'  ,'Ram'   ,'Scot' ,'Rose' ,np.nan],
     'age' : [25     , np.nan , 27     , 30    , 21    ,np.nan],
     'Sex' : ['M'    , 'M'    , np.nan , 'M'   , 'F'   ,np.nan],
    'Blank': [np.nan , np.nan , np.nan ,np.nan ,np.nan ,np.nan ]
    
})

print("DROP COLUMN: \n")
print("BEFORE:")
display(dropColumnData)
dropColumnData.dropna(axis = 'columns',how = 'all',inplace = True) 
print("AFTER 1: COLUMN WITH ALL ROWS AS NaN :")
display(dropColumnData)
dropColumnData.dropna(axis = 'columns',how = 'any',inplace = True) 
print("AFTER 2: COLUMN WITH ANY ROWS AS NaN :")
display(dropColumnData)
```
